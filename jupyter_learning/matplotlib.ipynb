# 画饼图

```python
%matplotlib inline
# 画图时嵌入输出
import matplotlib.pyplot as plt
labels = 'frog', 'hogs','dogs','logs'
sizes = 15,20, 45, 10
colors = 'yellowgreen', 'gold','lightskyblue','lightcoral'
explode = 0, 0.1, 0, 0 # 分开展示饼图
plt.pie(sizes, explode=explode, labels=labels, colors = colors, autopct='%1.1f%%', shadow=True, startangle =50)
plt.axis('equal')
plt.show()
```

# 绘图

```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

```python
# 创建一个8*6点的图，并设置分辨率80
plt.figure(figsize=(8,6), dpi = 80)

# 创建一个1*1的子图，接下来的图样会绘制在其中第一块
plt.subplot(1,1,1)

# 绘制余弦函数，使用蓝色的，连续的，宽度为1像素的线条
plt.plot(X, C, color="blue", linewidth=1.0, linestyle='-')

# 绘制正弦函数，使用绿色的，连续的，宽度为1像素的线条
plt.plot(X, S, color="green", linewidth=1.0, linestyle='-')

# 设置横轴的上下限
plt.xlim(-4.0, 4.0)

# 设置横轴坐标点
# endpoint 如果是真，则一定包括stop，如果为False，一定不会有stop
plt.xticks(np.linspace(-4,4,9,endpoint=True))

# 设置纵轴的上下限
plt.ylim(-1.0, 1.0)

# 设置纵轴坐标点
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# 在屏幕上显示
plt.show()
```

```python
# 创建一个8*6点的图，并设置分辨率80
plt.figure(figsize=(10,6), dpi = 80)

# 创建一个1*1的子图，接下来的图样会绘制在其中第一块
plt.subplot(1,1,1)

# 绘制余弦函数，使用蓝色的，连续的，宽度为1像素的线条
plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-')

# 绘制正弦函数，使用绿色的，连续的，宽度为1像素的线条
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-')

# 设置横轴的上下限
plt.xlim(-4.0, 4.0)

# 设置横轴坐标点
# endpoint 如果是真，则一定包括stop，如果为False，一定不会有stop
plt.xticks(np.linspace(-4,4,9,endpoint=True))

# 设置纵轴的上下限
plt.ylim(-1.0, 1.0)

# 设置纵轴坐标点
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# 在屏幕上显示
plt.show()
```

```python
# 创建一个8*6点的图，并设置分辨率80
plt.figure(figsize=(10,6), dpi = 80)

xmin, xmax = X.min(), X.max() 
ymin, ymax = C.min(), C.max()

dx = (xmax-xmin) * 0.2
dy = (ymax-ymin) * 0.2


# 创建一个1*1的子图，接下来的图样会绘制在其中第一块
plt.subplot(1,1,1)

# 绘制余弦函数，使用蓝色的，连续的，宽度为1像素的线条
plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-')

# 绘制正弦函数，使用绿色的，连续的，宽度为1像素的线条
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-')

# 设置横轴的上下限
plt.xlim(xmin-dx, xmax+dx)

# 设置横轴坐标点
# endpoint 如果是真，则一定包括stop，如果为False，一定不会有stop
plt.xticks(np.linspace(-4,4,9,endpoint=True))

# 设置纵轴的上下限
plt.ylim(ymin-dy, ymax+dy)

# 设置纵轴坐标点
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# 在屏幕上显示
plt.show()
```

```python
# 设置坐标点

# 创建一个8*6点的图，并设置分辨率80
plt.figure(figsize=(10,6), dpi = 80)

xmin, xmax = X.min(), X.max() 
ymin, ymax = C.min(), C.max()

dx = (xmax-xmin) * 0.2
dy = (ymax-ymin) * 0.2


# 创建一个1*1的子图，接下来的图样会绘制在其中第一块
plt.subplot(1,1,1)

# 绘制余弦函数，使用蓝色的，连续的，宽度为1像素的线条
plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-')

# 绘制正弦函数，使用绿色的，连续的，宽度为1像素的线条
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-')

# 设置横轴的上下限
plt.xlim(xmin-dx, xmax+dx)

# 设置横轴坐标点
# endpoint 如果是真，则一定包括stop，如果为False，一定不会有stop
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])

# 设置纵轴的上下限
plt.ylim(ymin-dy, ymax+dy)

# 设置纵轴坐标点
plt.yticks([-1,0,1])

# 在屏幕上显示
plt.show()
```

# 移动脊柱

```python
# 移动脊柱 spines

# 创建一个8*6点的图，并设置分辨率80
plt.figure(figsize=(10,6), dpi = 80)

xmin, xmax = X.min(), X.max() 
ymin, ymax = C.min(), C.max()

dx = (xmax-xmin) * 0.2
dy = (ymax-ymin) * 0.2


# 创建一个1*1的子图，接下来的图样会绘制在其中第一块
plt.subplot(1,1,1)

# 绘制余弦函数，使用蓝色的，连续的，宽度为1像素的线条
plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-', label='cosine')

# 绘制正弦函数，使用绿色的，连续的，宽度为1像素的线条
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-', label='sin')

# 设置横轴的上下限
plt.xlim(xmin-dx, xmax+dx)

# 设置横轴坐标点
# endpoint 如果是真，则一定包括stop，如果为False，一定不会有stop
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$','$-\pi/2$','$0$','$+\pi/2$','$+\pi$'])

# 设置纵轴的上下限
plt.ylim(ymin-dy, ymax+dy)

# 设置纵轴坐标点
plt.yticks([-1,0,1],['$-1$','$0$','$+1$'])

ax = plt.gca() # get current axis
ax.spines['right'].set_color('none') # 右侧不现实
ax.spines['top'].set_color('none')
# 设置坐标点在轴下侧
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # 设置底部轴的位置
# 设置坐标点在轴左侧
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0)) # 设置左侧轴的位置

# 增加图例
plt.legend(loc='upper left') # 展示label

# 在屏幕上显示
plt.show()
```

# 增加注释

```python
# 移动脊柱 spines

# 创建一个8*6点的图，并设置分辨率80
plt.figure(figsize=(10,6), dpi = 80)

xmin, xmax = X.min(), X.max() 
ymin, ymax = C.min(), C.max()

dx = (xmax-xmin) * 0.2
dy = (ymax-ymin) * 0.2


# 创建一个1*1的子图，接下来的图样会绘制在其中第一块
plt.subplot(1,1,1)

# 绘制余弦函数，使用蓝色的，连续的，宽度为1像素的线条
plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-', label='cosine')

# 绘制正弦函数，使用绿色的，连续的，宽度为1像素的线条
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-', label='sin')

# 设置横轴的上下限
plt.xlim(xmin-dx, xmax+dx)

# 设置横轴坐标点
# endpoint 如果是真，则一定包括stop，如果为False，一定不会有stop
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$','$-\pi/2$','$0$','$+\pi/2$','$+\pi$'])

# 设置纵轴的上下限
plt.ylim(ymin-dy, ymax+dy)

# 设置纵轴坐标点
plt.yticks([-1,0,1],['$-1$','$0$','$+1$'])

ax = plt.gca() # get current axis
ax.spines['right'].set_color('none') # 右侧不现实
ax.spines['top'].set_color('none')
# 设置坐标点在轴下侧
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # 设置底部轴的位置
# 设置坐标点在轴左侧
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0)) # 设置左侧轴的位置

# 增加图例
plt.legend(loc='upper left') # 展示label

t = 2*np.pi/3
plt.plot([t,t],[0, np.cos(t)], color='blue', linewidth=2.5, linestyle='--')
plt.scatter([t,],[np.cos(t)], 50, color = 'blue')
# 注释的点xy
# 注释的文本位置xytext
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t, np.cos(t)), xycoords='data',
            xytext=(-90, -50), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))

plt.scatter([t,],[np.sin(t)], 50, color = 'red')
# 注释的点xy
# 注释的文本位置xytext
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))

# 在屏幕上显示
plt.show()
```

# 3D图

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2, 2)
plt.show()
```
